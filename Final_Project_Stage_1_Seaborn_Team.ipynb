# **1. Decriptive Statistics**

Feature yang termasuk kategori data **Numerik**:

- **Administrative**: Jumlah halaman administratif yang dikunjungi oleh pengguna selama sesi tersebut.
- **Administrative Duration**: Durasi total yang dihabiskan oleh pengguna di halaman administratif selama sesi (dalam detik).
- **Informational**: Jumlah halaman informatif yang dikunjungi oleh pengguna selama sesi.
- **Informational Duration**: Durasi total yang dihabiskan oleh pengguna di halaman informatif selama sesi (dalam detik).
- **Product Related**: Jumlah halaman terkait produk yang dikunjungi oleh pengguna selama sesi.
- **Product Related Duration**: Durasi total yang dihabiskan oleh pengguna di halaman terkait produk selama sesi (dalam detik).
- **Bounce Rates**: Persentase pengunjung yang masuk ke halaman tersebut dan langsung meninggalkan situs web ("bounce") selama sesi.
- **Exit Rates**: Persentase keluar dari situs web setelah mengunjungi halaman tertentu selama sesi.
- **Page Values**: Nilai rata-rata dari halaman web yang pengguna kunjungi sebelum menyelesaikan transaksi e-commerce.

Feature yang termasuk kategori data **Kategorik**:
- **Special Day**: Indikator yang menunjukkan kedekatan waktu kunjungan situs dengan hari khusus tertentu (misalnya, Hari Ibu, Hari Valentine) di mana sesi lebih mungkin selesai dengan transaksi.
- **Month**: Bulan saat kunjungan dilakukan.
- **Operating Systems**:Sistem operasi yang digunakan oleh pengguna.
- **Browser**: Browser web yang digunakan oleh pengguna.
- **Region**: Wilayah geografis pengguna.
- **Traffic Type**: Jenis lalu lintas yang membawa pengguna ke situs web.
- **Visitor Type**: Tipe pengunjung, apakah pengunjung baru atau pengunjung yang kembali.
- **Weekend**: Indikator apakah kunjungan terjadi di akhir pekan atau tidak (True jika akhir pekan, False jika tidak).
- **Revenue**: Label kelas, indikator apakah pengguna menghasilkan pendapatan (True jika iya, False jika tidak).

```python
def calculate_missing_and_zero_values(df):
     # Calculate the number of zero values per column
    zero_values = (df == 0).sum()

    # Calculate the percentage of zero values
    zero_percentage = (zero_values / len(df)) * 100

    # Create a DataFrame to display the results
    missing_and_zero_data = pd.DataFrame({
        'Zero Values': zero_values,
        'Zero Percentage': zero_percentage
    })

    # Round the percentages to 2 decimal places
    missing_and_zero_data = missing_and_zero_data.round(2)

    return missing_and_zero_data

# Calculate and print the results
missing_and_zero_data = calculate_missing_and_zero_values(df)
print(missing_and_zero_data)
```

## Duplicates Data

```python
def calculate_duplicates(df):
    # Initialize dictionaries to hold the counts and percentages
    duplicate_counts = {}
    duplicate_percentages = {}

    # Iterate through each column in the DataFrame
    for column in df.columns:
        # Count the duplicate values (excluding the first occurrence)
        count = df.duplicated(subset=column).sum()

        # Calculate the percentage of duplicate values
        percentage = (count / len(df)) * 100

        # Store the results in the dictionaries
        duplicate_counts[column] = count
        duplicate_percentages[column] = percentage

    # Create a DataFrame to display the results
    duplicate_data = pd.DataFrame({
        'Duplicate Count': duplicate_counts,
        'Duplicate Percentage': duplicate_percentages
    })

    # Round the percentages to 2 decimal places
    duplicate_data = duplicate_data.round(2)

    return duplicate_data

# Calculate and print the results
duplicate_data = calculate_duplicates(df)
print(duplicate_data)
```

**INSIGHT:**
1. Tidak bisa dilakukan pengecekan baris duplikat karena tidak ada unique ID.

2. Duplikat disini menunjukkan nilai yang sama di masing-masing kolom, dibaca sebagai variasi.

## Data Outliers

```python
#pemisahan data categoricals & numericals
data = {
    'category': ['special_day', 'month', 'operating_systems', 'browser', 'region', 'traffic_type', 'visitor_type', 'weekend', 'revenue'],
    'numeric': ['administrative', 'administrative_duration', 'informational', 'informational_duration', 'product_related',
                'product_related_duration', 'bounce_rates', 'exit_rates', 'page_value']
}
```

```python
#check_outliers
def check_outliers(df, numeric_columns):
    outliers = pd.DataFrame(columns=df.columns)

    for col in numeric_columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[col] = (df[col] < lower_bound) | (df[col] > upper_bound)

    return df.assign(**outliers)

# Panggil fungsi check_outliers dengan df dan kolom numerik
numeric_columns = data['numeric']
df_with_outliers = check_outliers(df, numeric_columns)
print(df_with_outliers[numeric_columns].any())
```

```python
# Definisi fungsi untuk menghitung persentase outliers
def calculate_outlier_percentage(df, numeric_columns):
    outliers_df = check_outliers(df, numeric_columns)
    outlier_percentages = {}

    for col in numeric_columns:
        outlier_count = outliers_df[col].sum()
        total_count = len(outliers_df[col])
        outlier_percentages[col] = (outlier_count / total_count) * 100

    return outlier_percentages

# menghitung persentase outliers
numeric_columns = data['numeric']
outlier_percentages = calculate_outlier_percentage(df, numeric_columns)
outlier_percentages
```

Dari analisis outliers pada kolom numerik, ditemukan bahwa beberapa kolom memiliki persentase outliers yang signifikan:

Informational (21%)
Informational Duration (19%)
BounceRates (12%)
PageValues (22%)
Persentase outliers yang tinggi menunjukkan kemungkinan adanya kesalahan data, anomali, atau variabilitas yang sangat tinggi dalam data.

**Tindakan yang direkomendasikan meliputi pemeriksaan lebih lanjut, pembersihan data, transformasi data, atau penanganan outliers dalam model analisis.**

```python
# Menghitung jumlah data untuk setiap kombinasi SpecialDay, VisitorType_New_Visitor, dan Revenue
count_specialday_visitor = df.groupby(['special_day', 'visitor_type_New_Visitor'])['revenue'].value_counts().unstack(fill_value=0)

# Menghitung presentase Revenue 0 dan 1 untuk setiap kombinasi SpecialDay dan VisitorType_New_Visitor
prop_specialday_visitor = count_specialday_visitor.div(count_specialday_visitor.sum(axis=1), axis=0) * 100

# Plotting data dengan stack
visitor_types = df['visitor_type_New_Visitor'].unique()
special_days = df['special_day'].unique()

bar_width = 0.2
fig, ax = plt.subplots(figsize=(10, 6))

# Loop untuk setiap jenis VisitorType_New_Visitor
for i, visitor_type in enumerate(visitor_types):
    bottom = [0] * len(special_days)

    # Loop untuk setiap jenis Revenue
    for j, revenue_type in enumerate([0, 1]):
        counts = [prop_specialday_visitor.loc[(special_day, visitor_type), revenue_type] for special_day in special_days]
        bars = ax.bar([x + bar_width * i for x in range(len(special_days))], counts, bar_width, bottom=bottom, label=f'revenue {revenue_type} (visitor_type {visitor_type})')
        for bar, count in zip(bars, counts):
            ax.text(bar.get_x() + bar.get_width() / 2, bar.get_y() + bar.get_height() / 2, f'{count:.1f}%', ha='center', va='center')
        bottom = [b + c for b, c in zip(bottom, counts)]

# Menambahkan judul dan label
ax.set_xlabel('SpecialDay')
ax.set_ylabel('Percentage (%)')
ax.set_title('Percentage of Revenue by VisitorType_New_Visitor and SpecialDay')
ax.set_xticks([r + bar_width for r in range(len(special_days))])
ax.set_xticklabels(special_days)
ax.legend()

# Menampilkan plot
plt.show()
```

**Insight:**

- Untuk hari dimana tidak mendekati special day sama sekali, cenderung malah revenuenya lebih banyak. Dan dari yang terkonversi tersebut, 25% didukung dari Visitor baru (yang meskipun tidak ada tawaran apa apa atau tidak mendekati special day manapun) mereka menghasilkan revenue.

- Apabila dilihat dari special day 0.2 dan 0.4, ada revenue yang tinggi pada visitor baru. Padahal pada hari yang sangat mendekati special day(0.8 dan hari H special daynya (0.1) hanya mengonversi sebesar 11.1% dan 33.3%. Hal ini menunjukkan bahwa, visitor baru, terutama visitor baru yang terkonversi, tidak ada kaitanya (tidak karena terpengaruh oleh adanya special day.

- Tawaran atau campaign pada special day tidak mampu menggaet visitor atau pelanggan baru. Maka direkomendasikan untuk memperbaiki campaign special day tersebut.
